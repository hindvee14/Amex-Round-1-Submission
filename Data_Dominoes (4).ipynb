```python
# Calculate Venue Winning Rate
# Filter relevant columns
matches = match_lvl_data[['team1', 'team2', 'winner', 'venue']]

# Calculate the total matches played by each team at each venue
matches_played = matches.melt(id_vars=['venue'], value_vars=['team1', 'team2'], var_name='team_type', value_name='team')
matches_played = matches_played.groupby(['team', 'venue']).size().reset_index(name='matches_played')

# Calculate the total matches won by each team at each venue
matches_won = matches[matches['winner'].notna()]
matches_won = matches_won.groupby(['winner', 'venue']).size().reset_index(name='matches_won')

# Merge the two dataframes to calculate the win rate
win_rate = pd.merge(matches_played, matches_won, left_on=['team', 'venue'], right_on=['winner', 'venue'], how='left')
win_rate['matches_won'] = win_rate['matches_won'].fillna(0)
win_rate['win_rate'] = win_rate['matches_won'] / win_rate['matches_played']

# Prepare to merge with train_data
win_rate_team1 = win_rate.rename(columns={'team': 'team1', 'win_rate': 'team1_win_rate'})
win_rate_team2 = win_rate.rename(columns={'team': 'team2', 'win_rate': 'team2_win_rate'})

# Merge with train_data
train_data = pd.merge(train_data, win_rate_team1[['team1', 'venue', 'team1_win_rate']], on=['team1', 'venue'], how='left')
train_data = pd.merge(train_data, win_rate_team2[['team2', 'venue', 'team2_win_rate']], on=['team2', 'venue'], how='left')

# Merge with test_data
test_data = pd.merge(test_data, win_rate_team1[['team1', 'venue', 'team1_win_rate']], on=['team1', 'venue'], how='left')
test_data = pd.merge(test_data, win_rate_team2[['team2', 'venue', 'team2_win_rate']], on=['team2', 'venue'], how='left')

train_data.head()
test_data.head()
```

```python
# Function to update series standings and determine leading teams
def update_and_determine_leading_team(match_lvl_data):
    series_standings = {}
    team1_leading = []
    team2_leading = []

    for index, row in match_lvl_data.iterrows():
        series = row['series_name']
        team1 = row['team1']
        team2 = row['team2']
        winner = row['winner']

        if series not in series_standings:
            series_standings[series] = {team1: 0, team2: 0}

        if winner in series_standings[series]:
            series_standings[series][winner] += 1

        team1_wins = series_standings[series].get(team1, 0)
        team2_wins = series_standings[series].get(team2, 0)

        if team1_wins > team2_wins:
            team1_leading.append(1)
            team2_leading.append(0)
        elif team2_wins > team1_wins:
            team1_leading.append(0)
            team2_leading.append(1)
        else:
            team1_leading.append(0)
            team2_leading.append(0)

    match_lvl_data['team1_leading'] = team1_leading
    match_lvl_data['team2_leading'] = team2_leading

    return match_lvl_data

# Update the match_data DataFrame with leading team information
match_lvl_data = update_and_determine_leading_team(match_lvl_data)

# Merge the leading team information with the train_data
train_data = pd.merge(train_data, match_lvl_data[['match id', 'team1_leading', 'team2_leading']], on='match id', how='left')

# Merge the leading team information with the test_data
test_data = pd.merge(test_data, match_lvl_data[['match id', 'team1_leading', 'team2_leading']], on='match id', how='left')

train_data.head()
test_data.head()
```

Adding Features from Batsman Scorecard

```python
# Average Strike Rate of Each Inning
#Step 1: Calculate average strike rate for inning 1 and inning 2 for each match
avg_strike_rate_inning1 = batsman_lvl_data[batsman_lvl_data['inning'] == 1].groupby('match id')['strike_rate'].mean().reset_index()
avg_strike_rate_inning1.columns = ['match id', 'avg_strike_rate_inning1']

avg_strike_rate_inning2 = batsman_lvl_data[batsman_lvl_data['inning'] == 2].groupby('match id')['strike_rate'].mean().reset_index()
avg_strike_rate_inning2.columns = ['match id', 'avg_strike_rate_inning2']

# Step 2: Merge these average strike rates into the train data based on the match id
train_data = train_data.merge(avg_strike_rate_inning1, on='match id', how='left')
train_data = train_data.merge(avg_strike_rate_inning2, on='match id', how='left')

# Step 3: Merge these average strike rates into the test data based on the match id
test_data = test_data.merge(avg_strike_rate_inning1, on='match id', how='left')
test_data = test_data.merge(avg_strike_rate_inning2, on='match id', how='left')

train_data.head()
test_data.head()
```

```python
# Calculate total boundaries (fours + sixes) for each inning
batsman_lvl_data['total_boundaries'] = batsman_lvl_data['Fours'] + batsman_lvl_data['Sixes']

# Group by match_id and innings to get the total boundaries for each innings
boundaries_by_innings = batsman_lvl_data.groupby(['match id', 'inning'])['total_boundaries'].sum().unstack()

# Rename columns to match the required format
boundaries_by_innings.columns = ['total_boundaries_inning1', 'total_boundaries_inning2']

# Merge the total boundaries data with the train dataset based on match_id
train_data = train_data.merge(boundaries_by_innings, on='match id', how='left')

# Merge the total boundaries data with the test dataset based on match_id
test_data = test_data.merge(boundaries_by_innings, on='match id', how='left')

train_data.head()
test_data.head()
```

```python
#Total Runs scored by each team
# Calculate total runs scored by each team

inning1_runs = batsman_lvl_data[batsman_lvl_data['inning'] == 1].groupby('match id')['runs'].sum().reset_index()
inning1_runs.columns = ['match id', 'inning1_runs']

inning2_runs = batsman_lvl_data[batsman_lvl_data['inning'] == 2].groupby('match id')['runs'].sum().reset_index()
inning2_runs.columns = ['match id', 'inning2_runs']

# Step 2: Merge these average strike rates into the train data based on the match id
train_data = train_data.merge(inning1_runs, on='match id', how='left')
train_data = train_data.merge(inning2_runs, on='match id', how='left')

# Step 3: Merge these average strike rates into the test data based on the match id
test_data = test_data.merge(inning1_runs, on='match id', how='left')
test_data = test_data.merge(inning2_runs, on='match id', how='left')

train_data.head()
```

Features from Bowler Data

```python
# Economy Rate of Bowlers in each inning

ER_inning1 = bowler_lvl_data[bowler_lvl_data['inning'] == 1].groupby('match id')['economy'].mean().reset_index()
ER_inning1.columns = ['match id', 'ER_inning1']

ER_inning2 = bowler_lvl_data[bowler_lvl_data['inning'] == 2].groupby('match id')['economy'].mean().reset_index()
ER_inning2.columns = ['match id', 'ER_inning2']

# Merge the average economy rates into the train data
train_data = train_data.merge(ER_inning1, on='match id', how='left', suffixes=('_x', '_inning1'))
train_data = train_data.merge(ER_inning2, on='match id', how='left', suffixes=('_x', '_inning2'))

# Merge the average economy rates into the test data
test_data = test_data.merge(ER_inning1, on='match id', how='left', suffixes=('_x', '_inning1'))
test_data = test_data.merge(ER_inning2, on='match id', how='left', suffixes=('_x', '_inning2'))

train_data.head()
```

```python
#Bowling Average
bowler_lvl_data['BowlingAverage'] = bowler_lvl_data['runs'] / bowler_lvl_data['wicket_count']
bowl_avg_i1 = bowler_lvl_data[bowler_lvl_data['inning'] == 1].groupby('match id')['BowlingAverage'].mean().reset_index()
bowl_avg_i1.columns = ['match id', 'bowling_avg_i1']

bowl_avg_i2 = bowler_lvl_data[bowler_lvl_data['inning'] == 2].groupby('match id')['BowlingAverage'].mean().reset_index()
bowl_avg_i2.columns = ['match id', 'bowling_avg_i2']

# Merge into the train data
train_data = train_data.merge(bowl_avg_i1, on='match id', how='left')
train_data = train_data.merge(bowl_avg_i2, on='match id', how='left')

# Merge into the test data
test_data = test_data.merge(bowl_avg_i1, on='match id', how='left')
test_data = test_data.merge(bowl_avg_i2, on='match id', how='left')

train_data.head()
```

```python
#Bowler Strike Rate
bowler_lvl_data['StrikeRate'] = bowler_lvl_data['balls_bowled'] / (6 * bowler_lvl_data['wicket_count'])
bowl_strikerate_i1 = bowler_lvl_data[bowler_lvl_data['inning'] == 1].groupby('match id')['StrikeRate'].mean().reset_index()
bowl_strikerate_i1.columns = ['match id', 'bowling_strikerate_i1']

bowl_strikerate_i2 = bowler_lvl_data[bowler_lvl_data['inning'] == 2].groupby('match id')['StrikeRate'].mean().reset_index()
bowl_strikerate_i2.columns = ['match id', 'bowling_strikerate_i2']

# Merge into the train data
train_data = train_data.merge(bowl_strikerate_i1, on='match id', how='left', suffixes=('_x', '_i1'))
train_data = train_data.merge(bowl_strikerate_i2, on='match id', how='left', suffixes=('_x', '_i1'))

# Merge into the test data
test_data = test_data.merge(bowl_strikerate_i1, on='match id', how='left', suffixes=('_x', '_i1'))
test_data = test_data.merge(bowl_strikerate_i2, on='match id', how='left', suffixes=('_x', '_i1'))

train_data.head()
```

```python
#Dot Ball Percentage
bowler_lvl_data['DotBallPercentage'] = (bowler_lvl_data['dots'] / bowler_lvl_data['balls_bowled']) * 100
dotballp_i1 = bowler_lvl_data[bowler_lvl_data['inning'] == 1].groupby('match id')['DotBallPercentage'].mean().reset_index()
dotballp_i1.columns = ['match id', 'dotballp_i1']

dotballp_i2 = bowler_lvl_data[bowler_lvl_data['inning'] == 2].groupby('match id')['DotBallPercentage'].mean().reset_index()
dotballp_i2.columns = ['match id', 'dotballp_i2']

# Merge into the train data
train_data = train_data.merge(dotballp_i1, on='match id', how='left', suffixes=('_x', '_i1'))
train_data = train_data.merge(dotballp_i2, on='match id', how='left', suffixes=('_x', '_i1'))

# Merge into the test data
test_data = test_data.merge(dotballp_i1, on='match id', how='left', suffixes=('_x', '_i1'))
test_data = test_data.merge(dotballp_i2, on='match id', how='left', suffixes=('_x', '_i1'))

train_data.head()
```
